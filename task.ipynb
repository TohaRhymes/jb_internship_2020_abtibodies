```python
from RamachanDraw import fetch, phi_psi, plot
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

ID 3х веществ из банка [PDB](https://www.rcsb.org/):
* `2MGO` (oxytocin) - один из гормонов радости и счастья.
* `4BOR` - ацетилхолиновый рецептор (ацетилхолин тоже является гормоном, отвечающим за настроение).
* `1HSO`(HUMAN ALPHA ALCOHOL DEHYDROGENASE (ADH1A)) - фермент, катализирующий процесс окисления спиртов в крови до ацетальдегида. 
Нарисовать карту достаточно несложно, используя библиотеку `RamachanDraw`.

```python
def draw_and_save(PDB_id):
    # Нарисуем карту Рамачандрана
    data = fetch(PDB_id)
    plot(data, cmap='magma', save = True, show = False, out = PDB_id+"_map.png")
    # Сгенерируем словарь, чтобы хранить phi и psi
    # И сохраним проигнорированные аминокислотные остатки
    # Для данного задания это не необходимо, но может понадобиться в будущем
    phi_psi_dict, ignored_res = phi_psi(data, return_ignored=True)
    return phi_psi_dict, ignored_res
```

```python
# PDB id to be downloaded
PDB_id_oxy = '2MGO'
PDB_id_ace = '4BOR'
PDB_id_adh = '1HSO'
```

### Окситоцин

```python
%matplotlib inline
phi_psi_oxy, ignored_res_oxy = draw_and_save(PDB_id_oxy)
```

### Рецептор ацетилхолина

```python
phi_psi_ace, ignored_res_ace = draw_and_save(PDB_id_ace)
```

### Алкоголь дегидрогеназа

```python
phi_psi_adh, ignored_res_adh = draw_and_save(PDB_id_adh)
```
